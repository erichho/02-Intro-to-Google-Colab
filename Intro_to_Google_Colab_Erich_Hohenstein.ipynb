## 01 Importing Libraries

```python
import io
import pandas as pd
from google.colab import files
from google.colab import drive
```

## 01 Uploading your Data

### Direct Upload

```python
# Direct Upload
uploaded = files.upload()
```

```python
# Once the file finish uploading, we can read it into a dataframe
df = pd.read_csv(io.BytesIO(uploaded['netflix.csv']))
df.head(1)
```

### Reading directly from Google Drive
Another way to read all our files, is to mount our Google Drive. This means giving Google Colab access to our Google Drive project folder

```python
#This command will ask you to give persmisions to Google Colab to access your Google Drive files
drive.mount('/content/drive')
```

```python
#Once our Google Drive is mounted, we can read any file from it by giving its address
df = pd.read_csv('/content/drive/MyDrive/Blog/00 Data/netflix.csv')
df.head(1)
```
